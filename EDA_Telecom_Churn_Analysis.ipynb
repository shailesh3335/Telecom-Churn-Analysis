#### Chart - 2 - Pie Chart on Dependant Variable i.e., Churn (Univariate)

```python
# Dependant Column Value Counts
churn_counts = df['Churn'].value_counts()
print(churn_counts)
# Dependant Variable Column Visualization

plt.figure(figsize=(15, 6))
plt.pie(churn_counts, autopct="%1.1f%%", startangle=90, shadow=True, colors=['skyblue', 'red'], explode=[0, 0])

plt.legend(labels=['Not Churn (%)', 'Churn (%)'])
plt.title('Churn Distribution')

plt.show()
```

##### 3. Will the gained insights help creating a positive business impact? 
Are there any insights that lead to negative growth? Justify with specific reason.

Customer churn, also known as customer attrition or customer defection, refers to the rate at which customers are lost by a company. In the telecom industry, customer churn is a crucial business metric used to predict the number of customers who will discontinue their services with a telecom service provider. Churn is particularly significant in this industry as it directly impacts the competitiveness of the service provider.

While it is relatively easy to lose customers, acquiring new ones can be quite challenging. When a customer churns, it often leads to the loss of 3-4 potential customers who might have been acquired by the telecom service provider through positive word-of-mouth.

A notable example of this phenomenon occurred when Jio, a telecom provider, offered free internet services to everyone and rapidly acquired a large customer base. This resulted in an increased churn rate for other teleservice providers, as customers who switched to Jio shared the benefits of their experience with their relatives and friends, leading to further conversions. This illustrates the power of words, where negative or positive reviews can quickly spread and influence others' decisions.

In summary, customer churn poses a significant challenge in the telecom industry, impacting the competitiveness of service providers. Acquiring and retaining customers requires attention to both the factors that contribute to churn and the influence of word-of-mouth recommendations
